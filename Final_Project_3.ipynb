# Executive Summary

ShareCity is an ongoing research carried out by a team of passionate researchers base in Europe. They have an online database (SHARECITY100) which documents food sharing by ICT-enabled organisations across 100 cities around the world. The database is open-access, and the researchers' motivation for creating SHARECITY100 was to make visible the landscape of food sharing in cities by mapping it consistently across a large number of cities. See: http://sharecity.ie/

The database was downloaded for this project. There are 4000+ entries, with 50 entries for Singapore alone. Food sharing is an international movement and it is interesting to see what is happening in Asia and globally. What is being shared and how? How does cities in Asia compare with cities in other parts of the world?

This is an initial exploration of the dataset to shed light on what is shared and how, across the different cities, with a closer look at Asia. Clustering algorithms are initialised for 2 selected variables that show good correlation and clear differences in Asia cities and non-Asia cities, as an intial built towards better understanding and planning for targetted on-the-ground research.

# Risk and Assumptions

ShareCity Research is an ongoing project. Assumption is data is comprehensive and consistent enough to highlight trends in different cities.

# Hypothesis

Null hypothesis, $H_0$: Data is not useful to extract information on food sharing in Asia cities, and all 100 cities documented in ShareCity database

Alternative hypothesis, $H_1$: Analysis sheds light on what is shared and how, across the different cities, with a closer look if Asia cities would fall into clusters separate from non-Asia cities. Useful information can guide planning for targetted on-the-ground research.

```python
def Region(city):
    if city in Asia:
        return "Asia"
    elif city in Africa:
        return "Africa"
    elif city in Oceania:
        return "Oceania"
    elif city in Middle_East:
        return "Middle East"
    elif city in Americas:
        return "Americas"
    elif city in Europe:
        return "Europe"
    else:
        return "Others"
```

```python
def In_Asia(city):
    if city in Asia:
        return "Asia"
    else:
        return "Outside Asia"
```

```python
ShareCity_All['In_Asia'] = ShareCity_All['City'].map(In_Asia)
```

# Clustering

```python
# clustering is a unsupervised learning technique generally used to do initial profiling of data.
# it is helpful to sort out real-world data into groups based on similarity
# clusters should readily be visible when data is plotted

# in this clustering, we look at number of food organisations in a city vs diversity of what is shared
# we start first by making a duplicate of the dataframe on what is shared across the 100 cities
C_Share = ShareCity_What_Sum.copy()
```

```python
C_Share.head()
```

```python
# add a column to look at diversity of what is shared
C_Share["Share_Count"] = (C_Share.iloc[:,4:12] > 0).sum(1)
C_Share.head()
```

```python
# dataframe with numerical variables on what is shared, and whether a city is in/out of Asia
C_Share_AsiaIO = C_Share.drop(['Region', 'City'], axis=1)
C_Share_AsiaIO.head()
```

```python
C_Share_AsiaIO.info()
```

```python
C_Share_AsiaIO['Share_Compost'] = preprocessing.scale(C_Share_AsiaIO['Share_Compost'].astype('float64'))
C_Share_AsiaIO['Share_Food'] = preprocessing.scale(C_Share_AsiaIO['Share_Food'].astype('float64'))
C_Share_AsiaIO['Share_Kitchen'] = preprocessing.scale(C_Share_AsiaIO['Share_Kitchen'].astype('float64'))
C_Share_AsiaIO['Share_Knowledge'] = preprocessing.scale(C_Share_AsiaIO['Share_Knowledge'].astype('float64'))
C_Share_AsiaIO['Share_Land'] = preprocessing.scale(C_Share_AsiaIO['Share_Land'].astype('float64'))
C_Share_AsiaIO['Share_Meals'] = preprocessing.scale(C_Share_AsiaIO['Share_Meals'].astype('float64'))
C_Share_AsiaIO['Share_Plants'] = preprocessing.scale(C_Share_AsiaIO['Share_Plants'].astype('float64'))
C_Share_AsiaIO['Share_Seeds'] = preprocessing.scale(C_Share_AsiaIO['Share_Seeds'].astype('float64'))
C_Share_AsiaIO['Share_Skills'] = preprocessing.scale(C_Share_AsiaIO['Share_Skills'].astype('float64'))
C_Share_AsiaIO['Share_Count'] = preprocessing.scale(C_Share_AsiaIO['Share_Count'].astype('float64'))
```

```python
C_Share_AsiaIO.info()
```

```python
# As a start, 2 variables are chosen to initiate a cluster model
X = C_Share_AsiaIO[['Share_Knowledge', 'Share_Meals']]
X_scaled = StandardScaler().fit_transform( X )
```

```python
# finding the optimal number of clusters with a dendrogram
cmap = sns.cubehelix_palette(as_cmap=True, rot=-.3, light=1)
sns.clustermap(X_scaled, cmap=cmap, figsize=(15, 15))
```

Optimal number of clusters based on visual inspection of the dendrogram: 7 clusters. This is based on observing the maximum number of verticals that can be horizontally counted across the diagram, without going into the sub-divisions.

```python
# finding optimal number of clusters using elbow method
cluster_range = range( 1, 20 )
cluster_errors = []

for num_clusters in cluster_range:
  clusters = KMeans(num_clusters)
  clusters.fit(X)
  cluster_errors.append(clusters.inertia_)
```

```python
clusters_Xscaled = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
```

```python
clusters_Xscaled.head()
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Number of clusters')
plt.ylabel('Cluster errors')
plt.plot( clusters_Xscaled.num_clusters, clusters_Xscaled.cluster_errors, marker = "o" )
```

The elbow diagram above indicates that the gain in explained variance reduces significantly from 3 to 5. The optimal number of clusters could be 5. 

```python
# K-Means estimator algorithm
estimator = KMeans(n_clusters=5)
estimator.fit(X)
labels = estimator.labels_
print (labels)
centroids = estimator.cluster_centers_
```

```python
colors = set_colors(labels)
plt.figure(figsize=(12,10))
plt.xlabel("Share Knowledge")
plt.ylabel("Share Meals")
plt.scatter(X['Share_Knowledge'], X['Share_Meals'], c=colors)
```

```python
# DBSCAN estimator algorithm - can specify a minimum number of points and a minimum radius
estimator2 = DBSCAN(eps=0.2, min_samples=8)
estimator2.fit(X_scaled)
labels2 = estimator2.labels_
```

```python
colors2 = set_colors(labels2)
plt.figure(figsize=(12,10))
plt.xlabel("Share Knowledge")
plt.ylabel("Share Meals")
plt.scatter(X['Share_Knowledge'], X['Share_Meals'], c=colors2)
```

```python
# Agglomerative clustering
estimator3 = AgglomerativeClustering(n_clusters=5)
estimator3.fit(X_scaled)
labels3 = estimator3.labels_
```

```python
colors3 = set_colors(labels3)
plt.figure(figsize=(12,10))
plt.xlabel("Share Knowledge")
plt.ylabel("Share Meals")
plt.scatter(X['Share_Knowledge'], X['Share_Meals'], c=colors3)
```

```python
print ("Silhouette Coefficient for K-means (n=5): %0.3f" % metrics.silhouette_score(X_scaled, labels, metric='euclidean'))
print ("Silhouette Coefficient for DBScan (eps=0.2, min_samples=10): %0.3f" % metrics.silhouette_score(X_scaled, labels2, metric='euclidean'))
print ("Silhouette Coefficient for Agglomerative Clustering (n=5): %0.3f" % metrics.silhouette_score(X_scaled, labels3, metric='euclidean'))
```

The best value for silhouette is 1 and the worst value is -1. Given the metric scores, kmeans provide a better model for this dataset. Visually, the K-means and Agglomerative Clustering plots are consistent and have more visibly obvious clusters than the DSBSCAN plot. The clustering algorithm can still be improved, possibly with other variables that are not yet included in the model. It will also be interesting to include variables that are dissimilar.
